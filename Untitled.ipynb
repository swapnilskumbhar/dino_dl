```python
from fastai.vision.all import *
```

```python
import torch
```

```python
path_img = './imgs'
```

```python
dls = ImageDataLoaders.from_folder(Path(path_img),valid=None,bs=128,conver_mode='L',item_tfms=Resize((216,384),method='squish'))
```

```python
dls.show_batch()
```

```python
class custom_model(nn.Module):
    def __init__(self):
        super(custom_model, self).__init__()
        self.conv_layers = nn.Sequential(
        nn.Conv2d(3, 8, 7),
        nn.ReLU(),
        nn.BatchNorm2d(8),
        nn.Conv2d(8, 16, 7, stride=2),
        nn.ReLU(),
        nn.BatchNorm2d(16),
        nn.Conv2d(16, 32, 7, stride=2),
        nn.ReLU(),
        nn.BatchNorm2d(32),
        nn.Conv2d(32, 16, 7, stride=2),
        nn.ReLU(),
        nn.BatchNorm2d(16),
        nn.Conv2d(16, 8, 7, stride=2),
        nn.ReLU(),
        nn.Conv2d(8, 4, 7, stride=2),
        nn.ReLU()
        )
        self.linear = nn.Linear(24,3,bias=True)
        
    def forward(self, x):
        x = self.conv_layers(x)
        x = x.view(x.shape[0],-1)
        x = self.linear(x)
        return x
```

```python
learn = Learner(dls, custom_model(), metrics=error_rate, loss_func=torch.nn.CrossEntropyLoss())
```

```python
learn.lr_find()
```

```python
learn.fit_one_cycle(15,1e-3)
```

```python
learn.fit_one_cycle(10,1e-3)
```
